```python
# Показать текущую позицию игры
def show():
  print('  |', ' | '.join([str(i) for i in range(3)]))
  print('-'*14)
  row =''
  for i, j in enumerate(range(3)):
    row = field[i]
    print(str(i), '|' , ' | '.join(row))
    print('-'*14)

# Проверить корректность и сделать ход
def step(i, j, n):
  conditions = [i < 3, j < 3]
  if not all(conditions):
    return False
  if field[i][j] == ' ':
    field[i][j] = '0' if n%2 == 1 else 'x'
    show()
    return True
  return False

# Сообщить игроку об ошибке ввода
def input_error(gamer):
  print(f'Игрок "{gamer}", Ошибка ввода координат!')

# Проверить на наличие победителя
def finish_check(gamer):
  for path in win_paths:
    result = True
    for cell in path:
      if field[cell[0]][cell[1]] != gamer:
        result = False
        break
    if result:
      return True
  return False


# Ввод исходных данных
field = [[' ' for i in range(3)] for j in range(3)]
win_paths = [[(0, 0), (0, 1), (0, 2)], [(1, 0), (1, 1), (1, 2)], [(2, 0), (2, 1), (2, 2)], [(0, 0), (1, 0), (2, 0)], [(0, 1), (1, 1), (2, 1)], [(0, 2), (1, 2), (2, 2)], [(0, 0), (1, 1), (2, 2)], [(0, 2), (1, 1), (2, 0)]]

n, draw = 0, True

# Цикл игры
show()
while n < 9:
  # Определить игрока (начинают x)
  gamer = 'x' if n%2 == 0 else '0'
  i, j = input(f'Игрок "{gamer}", введите координаты хода через пробел: ').split()
  # Предварительная проверка введенных символов
  if not all([i.isdigit(), j.isdigit()]):
    input_error(gamer)
    continue
  # подготвока данных для осуществления хода
  *move, = int(i), int(j), n

  # Проверка корректности и ход
  if step(*move):
    # Проверка на наличие победителя
    if finish_check(gamer):
      print(f'Игрок "{gamer}" выиграл!')
      draw = False
      break
    n += 1
  else:
    input_error(gamer)

# завершение игры, если победитель был определен, то сообщение о ничье пропускается
if draw:
  print('Ничья!')
print('Игра окончена.')
```
